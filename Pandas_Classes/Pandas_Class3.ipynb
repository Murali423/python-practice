```python
#Merge Operation
data4 = {'emp_id': [101,102,103,104],
         'salary' : [234,345,200,400],
         'pf' : [23,34,56,78] 
         }

data5 = {'emp_id' : [101,102,104,105],
         'mob_no' : [3434,5656,767,878],
         'house_no' : [345,345,345,56]
         }

df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
pd.merge(df4,df5) #Basically performs inner join on Common Column (Emp_id)
```

```python
pd.merge(df4,df5,how='left')
```

```python
pd.merge(df4,df5,how='right')
```

```python
pd.merge(df4,df5,how='right',on='emp_id')
```

```python
data6 = {'emp_id1': [101,102,103,104],
         'salary' : [234,345,200,400],
         'pf' : [23,34,56,78] 
         }

data7 = {'emp_id2' : [101,102,104,105],
         'mob_no' : [3434,5656,767,878],
         'house_no' : [345,345,345,56]
         }

df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)
```

```python
pd.merge(df6,df7,left_on='emp_id1',right_on='emp_id2',how='inner')
```

```python
df_sales = pd.read_csv('C:\\Users\\HP\\Documents\\python-practice\\Pandas_Classes\\data fsds\\sales_data_final.csv')
df_sales
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a >= 0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
profit_flag(-45)
```

```python
profit_flag(4)
```

```python
df_sales.columns
```

```python
#Map,FIlter and Reduce function can be applied on lists but we want to apply such function for Series
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)  #Applying our function to each row of Dataframe
```

```python
df_sales.head()
```

```python
df_sales['len_customer_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a <10:
        return 'low'
    if a >=10 and a<20:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales.columns
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
df_sales
```
