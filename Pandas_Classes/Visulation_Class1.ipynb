```python
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
z = sin_fun(x,y)
```

```python
ax = plt.axes(projection= '3d')
ax.plot3D(x,y,z)
```

```python
ax = plt.axes(projection= '3d')
ax.plot3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
