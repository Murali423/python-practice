#### q3 : Try to print all the number divisible by 3 in between a range of 40 - 400

```python
a  = 40
b = 400
li = []
while a <= b:
    if a % 3 == 0 :
        li.append(a)
    a += 1
print(li)
```

#### q17 : write a function whihc will take multiple list as a input and give me concatnation of all the element as and output

```python
def cn_list(l,l1,l2,l3):
    m = [l,l1,l2,l3]
    if type(l) == list:
        for i in m[1:]:
            l.extend(i)
    return l

l = [1,2,3,4]
l1 = [5,6,7,9]
l2 = [8,9,3,4]
l3 = [3,6,2,1]

res = cn_list(l,l1,l2,l3)
print('list of functions: ',res)
```
