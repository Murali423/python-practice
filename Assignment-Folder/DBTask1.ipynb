#### 6. In SQL task, try to perform left join operation with Attribute dataset and dress dataset on column Dress_ID

```python
import mysql.connector as conn
import pandas as pd
import logging
```

```python
try:
    logging.debug('creating connection with database')
    mydb = conn.connect(host='localhost', user = 'root', password = 'MuR@li@0423', database ='task')
    logging.info('Connection to mysql db is successful')
    logging.debug('creating the datframe pandas from sql')
    query = 'select * from attribute_dataset as a left join Dress_Sales as d on a.dress_id = d.dress_id;'
    lj = pd.read_sql_query(query,mydb)
    lj_df = pd.DataFrame(lj)
    logging.info('Left join dataframe created successfully')
except Exception as e:
    logging.exception(e)
finally:
    mydb.close()
```

#### 7. Write the SQL query to find out how many unique dress that we have based on Dress_ID

```python


mydb = conn.connect(host='localhost', user = 'root', password = 'MuR@li@0423', database ='task')

query = 'select count(distinct(dress_id)) from attribute_dataset ;'
cursor = mydb.cursor()
cursor.execute(query)
for i in cursor.fetchall():
    print("Unique dress count is ",i[0])
mydb.close()
```
